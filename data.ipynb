```python
df = df.drop([
        'Unnamed: 0', 'address', 'AdUpdateDate', 'Category',
        'UsingStatus', 'EligibilityForInvestment', 'BuildingAge',
       'BuildStatus', 'TitleStatus', 'NumberOfWCs', 
       'AdCreationDate', 'Type', 'NetSquareMeters', 'HeatingType',
       'CreditEligibility', 'InsideTheSite', 'StructureType', 'MortgageStatus', 
       'Swap', 'Balcony', 'PriceStatus', 'RentalIncome', 'NumberOfBalconies', 
       'BalconyType', 'HallSquareMeters', 'WCSquareMeters', 'IsItVideoNavigable?',
       'Subscription', 'BathroomSquareMeters', 'BalconySquareMeters'
], axis='columns')
df
```

```python
df['GrossSquareMeters'] = df['GrossSquareMeters'].str.replace(' m2', '')
df['GrossSquareMeters'] = pd.to_numeric(df['GrossSquareMeters'])
df
```

```python
sns.boxplot(df['price'])
```

```python
df = df[df['price'] <= 4000000]
sns.boxplot(df['price'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
print("Price min, max: ", df['price'].min(), df['price'].max())
print("GSM min, max: ", df['GrossSquareMeters'].min(), df['GrossSquareMeters'].max())
scaler = MinMaxScaler(feature_range=(0, 1))
df[['price', 'GrossSquareMeters']] = scaler.fit_transform(df[['price', 'GrossSquareMeters']])
df
```

```python
df.isnull().sum()
```

```python
dist_rate = df.ItemStatus.value_counts(normalize = True)
missing = df['ItemStatus'].isnull()
df.loc[missing, 'ItemStatus'] = np.random.choice(dist_rate.index, size = len(df[missing]), p = dist_rate.values)
df
```

```python
df.isnull().sum()
```

```python
df['NumberOfRooms'] = df['NumberOfRooms'].replace({'8+ Oda': '8+', 'Stüdyo': '1', '1 Oda': '1', '2.5+1': '5+1', '3.5+1': '5+1', '1.5+1': '5+1', '4.5+1': '5+1', '5 Oda': '5'})
df['NumberOfRooms'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}
mappings = {}

encoders['ItemStatus'] = LabelEncoder()
df['ItemStatus'] = encoders['ItemStatus'].fit_transform(df['ItemStatus'])
mappings['ItemStatus'] = {code: name for code, name in enumerate(encoders['ItemStatus'].classes_)}

encoders['NumberOfRooms'] = LabelEncoder()
df['NumberOfRooms'] = encoders['NumberOfRooms'].fit_transform(df['NumberOfRooms'])
mappings['NumberOfRooms'] = {code: name for code, name in enumerate(encoders['NumberOfRooms'].classes_)}

encoders['FloorLocation'] = LabelEncoder()
df['FloorLocation'] = encoders['FloorLocation'].fit_transform(df['FloorLocation'])
mappings['FloorLocation'] = {code: name for code, name in enumerate(encoders['FloorLocation'].classes_)}

encoders['district'] = LabelEncoder()
df['district'] = encoders['district'].fit_transform(df['district'])
mappings['district'] = {code: name for code, name in enumerate(encoders['district'].classes_)}

for key, mapping in mappings.items():
    print(f'{key} mapping:', mapping)
```

```python
df['NumberOfBathrooms'] = df['NumberOfBathrooms'].replace({'Yok': 0, '6+': 6})
df['NumberOfBathrooms'] = df['NumberOfBathrooms'].astype(int)
```

```python
df.to_csv('./datasets/model_input_dataset.csv', index=False)
```
